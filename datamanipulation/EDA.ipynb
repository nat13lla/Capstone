```python
# Heatmap to show the correlation between the columns.
plt.figure(figsize = (12, 12))
sns.heatmap(df.corr(numeric_only=True),
            annot = True,
            fmt = '.2',
            cmap = 'vlag',
            mask = np.triu(df.corr(numeric_only=True)),
            vmin = -1,
            vmax = 1)

plt.show()
```
