```python
# Imports:
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read in the final_dataframe.csv from github
df = pd.read_csv("https://raw.githubusercontent.com/nat13lla/Capstone/main/data_files/final_dataframe.csv")
```

## Data Cleaning

### Combine name columns

```python
# Combine first and second name to create a driver name column
df['driver_name'] = df['givenName']+' '+df['familyName']
# Drop the other columns (givenName, familyName)
df = df.drop(['givenName','familyName' ], axis = 1)
```

```python
df.head()
```

### Re-order Columns

```python
# Look at all Column Titles
df.columns
```

```python
# Reorder
df = df[['results_id', 'Season', 'Round', 'Race Name', 'race_id', 'date', 'time',
         'Circuit Name','circuit_id', 'location', 'country', 'lat', 'lng',
         'drivers_id', 'driverId','driver_name', 'code','perm_number', 'nationality','dateOfBirth',
         'constructors_id', 'constructorId','constructorname', 'team_nationality',
         'grid','position', 'points', 'status', 'laps',
         't_millis', 't_time', 'fl_av_speed', 'fl_time', 'fl_lap_no', 'fl_rank']]
```

```python
df.head()
```

### Rename Columns

```python
# rename the columns
df.rename(columns={'Season':'year','Round':'round','Race Name':'race_name',
                   'date':'date_of_race', 'time':'time_of_race','Circuit Name':'circuit_name',
                   'nationality':'driver_nationality','dateOfBirth':'dob','constructorname':'constructor_name','team_nationality':'constructor_nationality',
                   'grid':'start_position', 'position':'end_position','points':'points_gained',
                   'status':'race_status','laps':'laps_done',
                   't_millis':'race_millis','t_time':'race_time'},inplace=True)
```

```python
df.head()
```

### Add places_gained column

```python
# Add places Gained Column
df['places_gained'] = df['start_position']-df['end_position']
df.head(10)
```

### Removing columns:
Removing the ID columns
* 'results_id', 'race_id', 'circuit_id', 'driver_id', 'constructor_id'
* 'constructorId', 'driverId'


```python
# Remove the ID columns
df = df.drop(['results_id', 'race_id', 'circuit_id', 'drivers_id', 'constructors_id' ], axis = 1)
```

```python
# Remove alternate id columns
df = df.drop(['constructorId', 'driverId'], axis = 1)
```

* Remove lat/lng
  - as gives same data as location/country

```python
# Remove lat/lng
df = df.drop(['lat', 'lng'], axis = 1)
```
