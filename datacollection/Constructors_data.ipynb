# Data Collection

```python
# Imports:
import pandas as pd
import numpy as np
import requests # default library for talking to the web
from pprint import pprint # 'Pretty Print': prints less offensive JSON
```

# Constructors 2022

```python
# Connect to the API
url22 = 'http://ergast.com/api/f1/2022/constructors.json'
# Get a response
response = requests.get(url22)
print(response)

# 200 - Life is good :)
# 404 - Life is bad :(
# 400 - Your request is nonsense
```

```python
# Data output
data22 = response.json()

pprint(data22)
```

```python
# Create a dataframe containing all the data
constructors_data22 = data22['MRData']['ConstructorTable']['Constructors']
constructors_22 = pd.DataFrame(constructors_data22)

# Show
constructors_22
```

```python
# Download the Dataframe as CSV
constructors_22.to_csv('constructors_22.csv', index=False)
```

# Combine into 1

```python
# Concatenate the drivers from all 3 seasons
combined_constructors = pd.concat([constructors_22, constructors_23, constructors_24])
combined_constructors
```

```python
# Drop the Duplicated Drivers
combined_constructors = combined_constructors.drop_duplicates()
combined_constructors.reset_index(drop=True, inplace=True)
combined_constructors
```

* rb is the rebrand of AlphaTauri
* Sauber is the rebrand of Alfa Romeo

```python
# Add in a circuit id column
constructors_id = combined_constructors.copy()  # Make a copy
constructors_id['constructors_id'] = combined_constructors.index + 1  # Add column id

# Set as index
constructors_id.set_index('constructors_id', inplace=True)

# Show
constructors_id
```

```python
# Download the Dataframe as CSV
constructors_id.to_csv('constructors_all.csv', index=True)
```
