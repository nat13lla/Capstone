# Data Collection

```python
# Imports:
import pandas as pd
import numpy as np
import requests # default library for talking to the web
from pprint import pprint # 'Pretty Print': prints less offensive JSON
```

# 2022

```python
results22 = []
def get_details(race_number):
    try:
        url22 = f'http://ergast.com/api/f1/2022/{race_number}/results.json'
        response = requests.get(url22) # connect to site
        data22 = response.json()
        results_info22 = data22['MRData']['RaceTable']['Races'][0]

        for result in results_info22['Results']:
            driver = result['Driver']
            times = result.get('Time', {})
            flap = result.get('FastestLap', {})
            constructor = result['Constructor']

            results_data22 = {
                'Season': results_info22['season'],
                'Round': results_info22['round'],
                'Race Name': results_info22['raceName'],
                'Circuit Name': results_info22['Circuit']['circuitName'],
                'driverId': driver['driverId'],
                'givenName': driver['givenName'],
                'familyName': driver['familyName'],
                'perm_number': driver['permanentNumber'],
                'code': driver['code'],
                'constructorId': constructor['constructorId'],
                'constructorname': constructor['name'],
                'nationality': driver['nationality'],
                'position': result['position'],
                'points': result['points'],
                'status': result['status'],
                'position': result['position'],
                'grid': result['grid'],
                'laps':result['laps'],
                't_millis': times.get('millis'),
                't_time': times.get('time'),
                'fl_av_speed':flap.get('AverageSpeed',{}).get('speed'),
                'fl_time':flap.get('Time', {}).get('time'),
                'fl_lap_no':flap.get('lap'),
                'fl_rank':flap.get('rank')

            }
            results22.append(results_data22)

    except Exception as e:
        print(f"Error: {e}")

```

```python
for race_number in range(1, 23):
    get_details(race_number)

results_22 = pd.DataFrame(results22)



results_22
```

```python
# Download the Dataframe as CSV
results_22.to_csv('results_22.csv', index=False)
```
