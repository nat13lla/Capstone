```python
import pandas as pd
import numpy as np
```

# Combining the dataframes

```python
# Remove duplicate racename column
races = races.drop(['name'], axis = 1)
races.head()
```

## Race and Results

```python
merged_data = pd.merge(results, races, how='inner', left_on=['Race Name', 'Season', 'Round'], right_on=['raceName', 'year', 'round'])

# Drop redundant columns if needed
merged_data.drop(columns=['year', 'round', 'raceName'], inplace=True)
```

```python
merged_data
```

## + Drivers

```python
merged_data.columns
```

```python
drivers.head()
```

```python
drivers.columns
```

```python
# Drop the 'url' column
drivers.drop(columns=['url','permanentNumber', 'code', 'givenName', 'familyName', 'nationality'], inplace=True)
```

```python
drivers.head()
```

```python
# Merge the data
merged_data2 = pd.merge(merged_data, drivers, how='inner', on=['driverId'])
```

```python
merged_data2
```

## + constructors

```python
merged_data2.columns
```

```python
constructors.head()
```

```python
constructors.columns
```

```python
# Drop the 'url' column
constructors.drop(columns=['url', 'name'], inplace=True)
```

```python
constructors = constructors.rename(columns={'nationality': 'team_nationality'})
```

```python
constructors.head()
```

```python
# Merge the data
merged_data3 = pd.merge(merged_data2, constructors, how='inner', on=['constructorId'])
```

```python
merged_data3
```

# + circuit

```python
merged_data3.columns
```

```python
circuit.head()
```

```python
circuit.columns
```

```python
# Drop the 'url' column
circuit.drop(columns=['circuit'], inplace=True)
```

```python
circuit = circuit.rename(columns={'raceName': 'Race Name'})
```

```python
circuit.head()
```

```python
# Merge the data
merged_data4 = pd.merge(merged_data3, circuit, how='inner', on=['Race Name'])
```

```python
merged_data4
```

```python
final_df = merged_data4.sort_values(by='results_id')
```

```python
final_df
```

```python
final_df.reset_index(drop=True, inplace=True)
final_df
```

```python
# Download the Dataframe as CSV
final_df.to_csv('final_dataframe.csv', index=False)
```
