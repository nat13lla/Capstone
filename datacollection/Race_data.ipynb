# Data Collection

```python
# Imports:
import pandas as pd
import numpy as np
import requests # default library for talking to the web
from pprint import pprint # 'Pretty Print': prints less offensive JSON
```

# Races 2022

```python
# Connect to the API
url22 = 'http://ergast.com/api/f1/2022.json'
# Get a response
response = requests.get(url22)
print(response)

# 200 - Life is good :)
# 404 - Life is bad :(
# 400 - Your request is nonsense
```

```python
# Data output
data22 = response.json()

pprint(data22)
```

```python
races_data22 = data22['MRData']['RaceTable']['Races']

# Take relevant circuit details from each race that year
races22 = []
for race in races_data22:
    race_info = {
        'year': race['season'],
        'round': race['round'],
        'name': race['raceName'],
        'date': race['date'],
        'time': race['time'],
        'raceName': race['raceName']
    }
    races22.append(race_info)

# Create a dataframe with the extracted data
race_22 = pd.DataFrame(races22)

# Show
race_22
```

```python
# Download the Dataframe as CSV
race_22.to_csv('race_22.csv', index=False)
```
