# Data Collection

```python
# Imports:
import pandas as pd
import numpy as np
import requests # default library for talking to the web
from pprint import pprint # 'Pretty Print': prints less offensive JSON
```

#Drivers Standings 2022

```python
# Connect to the API
url22 = 'http://ergast.com/api/f1/2022/driverStandings.json'
# Get a response
response = requests.get(url22)
print(response)

# 200 - Life is good :)
# 404 - Life is bad :(
# 400 - Your request is nonsense
```

```python
# Data output
data22 = response.json()

pprint(data22)
```

```python
# Construcor info
# # Extracting relevant data
# standings = data22['MRData']['StandingsTable']['StandingsLists']
# # Take relevant standings details from each race that year
# driver_standings = []
# for standings_list in standings:
#     for driver_info in standings_list['DriverStandings']:
#         constructor_info = driver_info['Constructors'][0]
#         driver_standings.append({
#             'Constructor ID': constructor_info['constructorId'],
#             'Constructor Name': constructor_info['name'],
#             'Constructor Nationality': constructor_info['nationality'],
#             'Constructor URL': constructor_info['url'],
#         })
```

```python
# Extracting relevant data
standings = data22['MRData']['StandingsTable']['StandingsLists']

driver_standings = []
for standings_list in standings:
    for driver_info in standings_list['DriverStandings']:
        constructor_info = driver_info['Constructors'][0]
        driver_standings.append({
            'Driver Code': driver_info['Driver']['code'],
            'Driver Permanent Number': driver_info['Driver']['permanentNumber'],
            'Constructor ID': constructor_info['constructorId'],
            'Points': driver_info['points'],
            'Position': driver_info['position'],
            'Wins': driver_info['wins']
        })

# Creating DataFrame
standings_22 = pd.DataFrame(driver_standings)

standings_22
```

```python
# Download the Dataframe as CSV
standings_22.to_csv('standings_22.csv', index=False)
```
