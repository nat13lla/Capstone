# Data Collection

```python
# Imports:
import pandas as pd
import numpy as np
import requests # default library for talking to the web
from pprint import pprint # 'Pretty Print': prints less offensive JSON
```

# Circuits 2022

```python
# Connect to the API
url22 = 'http://ergast.com/api/f1/2022.json'
# Get a response
response = requests.get(url22)
print(response)

# 200 - Life is good :)
# 404 - Life is bad :(
# 400 - Your request is nonsense
```

```python
# Data output
data22 = response.json()
pprint(data22)
```

```python
races_data22 = data22['MRData']['RaceTable']['Races']

# Take relevant circuit details from each race that year
circuits22 = []
for race in races_data22:
    circuit_info = race['Circuit']
    circuit_data22 = {
        'circuit': circuit_info['circuitName'].lower().replace(" ", "_"),
        'location': circuit_info['Location']['locality'],
        'country': circuit_info['Location']['country'],
        'lat': circuit_info['Location']['lat'],
        'lng': circuit_info['Location']['long'],
        'raceName': race['raceName']
    }
    circuits22.append(circuit_data22)

# Create a dataframe with the extracted data
circuits_22 = pd.DataFrame(circuits22)

# Show
circuits_22
```

```python
# Download the Dataframe as CSV
circuits_22.to_csv('circuits_22.csv', index=False)
```

# Combine into 1

```python
# Concatenate the circuits from all 3 seasons
combined_circuits = pd.concat([circuits_22, circuits_23, circuits_24])
combined_circuits
```

```python
# Drop the Duplicate Circuits
combined_circuits = combined_circuits.drop_duplicates()
combined_circuits.reset_index(drop=True, inplace=True)
combined_circuits
```

```python
# Add in a circuit id column
circuits_id = combined_circuits.copy()  # Make a copy
circuits_id['circuit_id'] = combined_circuits.index + 1  # Add column id

# Set as index
circuits_id.set_index('circuit_id', inplace=True)

# Show
circuits_id
```

```python
# Download the Dataframe as CSV
circuits_id.to_csv('circuits_all.csv', index=True)
```
