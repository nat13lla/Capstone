# Data Collection

```python
import pandas as pd
import numpy as np
import requests # default library for talking to the web
from pprint import pprint # 'Pretty Print': prints less offensive JSON
```

# Circuits 2022

```python
# List of all circuits within 2022 season
url22 = 'http://ergast.com/api/f1/2022.json'
#
response = requests.get(url22)
print(response)

# 200 - Life is good :)
# 404 - Life is bad :(
# 400 - Your request is nonsense
```

```python
data22 = response.json()
pprint(data22)
```

```python
races_data22 = data22['MRData']['RaceTable']['Races']

# Extracting circuit details from each race
circuits22 = []
for race in races_data22:
    circuit_info = race['Circuit']
    circuit_data22 = {
        'circuitRef': circuit_info['circuitName'].lower().replace(" ", "_"),
        'location': circuit_info['Location']['locality'],
        'country': circuit_info['Location']['country'],
        'lat': circuit_info['Location']['lat'],
        'lng': circuit_info['Location']['long'],
        'raceName': race['raceName']
    }
    circuits22.append(circuit_data22)

# Creating a DataFrame from the extracted circuit data
circuits_22 = pd.DataFrame(circuits22)

# Displaying the DataFrame
circuits_22
```

```python
circuits_22.to_csv('circuits_22.csv', index=False)
```
