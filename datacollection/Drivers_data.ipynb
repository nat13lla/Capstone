# Data Collection

```python
# Imports:
import pandas as pd
import numpy as np
import requests # default library for talking to the web
from pprint import pprint # 'Pretty Print': prints less offensive JSON
```

#Drivers 2022

```python
# Connect to the API
url22 = 'http://ergast.com/api/f1/2022/drivers.json'
# Get a response
response = requests.get(url22)
print(response)

# 200 - Life is good :)
# 404 - Life is bad :(
# 400 - Your request is nonsense
```

```python
# Data output
data22 = response.json()

pprint(data22)
```

```python
# Create a dataframe containing all the data
drivers_data22 = data22['MRData']['DriverTable']['Drivers']
drivers_22 = pd.DataFrame(drivers_data22)

# Show
drivers_22
```

```python
# Download the Dataframe as CSV
drivers_22.to_csv('drivers_22.csv', index=False)
```

# Drivers 2023:

# Drivers 2024:

```python
# Connect to the API
url24 = 'http://ergast.com/api/f1/2024/drivers.json'
# Get a response
response = requests.get(url24)
print(response)

# 200 - Life is good :)
# 404 - Life is bad :(
# 400 - Your request is nonsense
```

# Combine into 1

```python
# Concatenate the drivers from all 3 seasons
combined_drivers = pd.concat([drivers_22, drivers_23, drivers_24])
combined_drivers
```

```python
# Drop the Duplicated Drivers
combined_drivers = combined_drivers.drop_duplicates()
combined_drivers.reset_index(drop=True, inplace=True)
combined_drivers
```

```python
# Download the Dataframe as CSV
combined_drivers.to_csv('drivers_all.csv', index=False)
```
